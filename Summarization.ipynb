```python
#menyimpan text yang akan di ringkas ke dalam satu vareabel

text = '''
Kerajaan Majapahit adalah salah satu kerajaan terbesar yang pernah ada di Indonesia. Berdiri pada abad ke-14, kerajaan ini berpusat di wilayah Jawa Timur, tepatnya di kota Trowulan. Kerajaan Majapahit terkenal dengan kekuatannya yang luar biasa, serta perkembangan budaya dan ilmu pengetahuan yang terjadi di wilayahnya.

Pada awalnya, Kerajaan Majapahit merupakan sebuah kerajaan kecil yang terletak di daerah selatan Jawa. Namun, berkat kepemimpinan yang tangguh dan strategi yang cerdas, kerajaan ini berhasil tumbuh menjadi salah satu kerajaan terbesar di Asia Tenggara pada abad ke-14.

Kerajaan Majapahit dibangun oleh seorang raja bernama Raden Wijaya, yang merupakan putra dari Raja Kertanegara, raja terakhir dari Kerajaan Singhasari. Raden Wijaya memperluas wilayah kerajaannya dengan memanfaatkan kelemahan yang dialami oleh Kerajaan Singhasari, yang sedang terpuruk pada saat itu.

Setelah berhasil memperluas wilayahnya, Raden Wijaya memindahkan pusat kerajaannya ke Trowulan, yang kemudian menjadi ibukota Kerajaan Majapahit. Pada awalnya, Kerajaan Majapahit hanya terdiri dari wilayah Jawa Timur, namun seiring dengan perkembangan zaman, wilayah kerajaan ini semakin luas dan merambah hingga ke wilayah lain di Indonesia, seperti Bali, Lombok, dan Kalimantan.

Kerajaan Majapahit memiliki sistem pemerintahan yang terorganisir, dengan raja sebagai kepala negara dan memiliki beberapa mentri yang bertanggung jawab atas berbagai bidang, seperti perdagangan, pertanian, dan militer. Selain itu, Kerajaan Majapahit juga terkenal dengan kemajuan yang dicapainya dalam bidang seni, sastra, dan ilmu pengetahuan.

Majapahit juga terkenal dengan perdagangan luar negeri yang merajalela, yang membantu memperluas pengaruh kerajaan tersebut ke wilayah lain di Asia Tenggara. Kerajaan Majapahit juga memiliki hubungan yang erat dengan kerajaan-kerajaan lain di wilayah Indonesia, seperti Kerajaan Sunda dan Kerajaan Bali.

Pada akhir abad ke-16, Kerajaan Majapahit mulai mengalami penurunan kekuatan
'''
```

```python
#memecah kalimat berdasarkan titik
import re
text =text.replace('\n', '')
sentence = re.split('\. |\.',text)
sentence
```

---memisahkan kalimat menjadi perkata---
(Tokenizer)

```python
len(sentence)
```

```python
import nltk
nltk.download('punkt')
```

```python
tokenizer = nltk.RegexpTokenizer(r"\w+")
tokenized = [tokenizer.tokenize(s.lower()) for s in sentence]
tokenized
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
listStopword = set(stopwords.words('indonesian'))
listStopword
```

--menghapus kata yang dianggap tidak penting--




```python
important_token = []
for sent in tokenized:
  filtered = [s for s in sent if s not in listStopword]
  important_token.append(filtered)
important_token
```

```python
sw_removed = [' '.join(t) for t in important_token]
sw_removed
```

--mengembalikan kata menjadi kata dasar -- 

```python
#install package Sastrawi
!pip install Sastrawi
```

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
factory = StemmerFactory()
stemmer = factory.create_stemmer()
```

```python
stemmed_sent = [stemmer.stem(sent) for sent in sw_removed]
stemmed_sent
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vec = TfidfVectorizer(lowercase=True)
document = vec.fit_transform(stemmed_sent)
```

```python
document = document.toarray()
```

```python
document.shape
```

```python
n = 5
```

```python
import numpy as np
```

```python
result = np.sum(document, axis=1)
result.shape
```

```python
sorted(result)
```

```python
top_n = np.argsort(result)[-n:]
top_n
```

```python
sum_index = sorted(top_n)
sum_index
```

```python
print('hasil ringkasan: ')
for i in sum_index:
  print(sentence[i], end='.\n')
```
